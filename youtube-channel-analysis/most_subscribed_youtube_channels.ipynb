## Upload Dataset

- To begin the analysis, I upload the dataset using Google Colab's file upload feature.

- The dataset should be in CSV format and contain information about YouTube channels such as subscribers, views, and categories.

- Once uploaded, the file will be loaded into a pandas DataFrame for further analysis.



```python
import io
import pandas as pd
import plotly.express as px


df = pd.read_csv('data/most_subscribed_youtube_channels.csv')


df['subscribers'] = df['subscribers'].str.replace(',', '', regex=False).astype(float)

display(df.head())
```

## Loading and Preparing the Data

- After uploading the dataset, I load it into a pandas DataFrame for analysis.

- The 'subscribers' column contains values formatted with commas (e.g., "1,000,000"), which prevents it from being treated as a numeric variable.  
To fix this, I remove the commas and convert the column to a numeric type.

- Finally, I display the first few rows to verify that the data has been loaded and cleaned correctly.



---



## Subscribers Distribution

- In this section, I analyze the distribution of subscriber counts among the most subscribed YouTube channels.

- A histogram is used to visualize how subscriber values are spread and to identify patterns such as concentration or skewness.

```python
fig = px.histogram(df, x='subscribers', title='Subscriber Count')
fig.show()
```

### **Insight**

- From the histogram, we can observe that subscriber counts are not evenly
distributed.

- Most channels fall within a certain range, while a smaller number of channels have extremely high subscriber counts.  
- This suggests a ***skewed distribution*** where a few top channels dominate in terms of subscribers.

- This pattern is common in platforms like YouTube, where a small number of creators attract a large portion of the audience.

## Category Distribution

- In this section, I analyze how subscribers are distributed across different YouTube categories.

- A pie chart is used to show the proportion of total subscribers that each category represents.

```python
fig = px.pie(df, values='subscribers', names='category', title='YouTube Categories')
fig.show()
```

### Insight

- From the pie chart, we can observe that certain categories dominate in terms of total subscribers.

- Some categories occupy a larger portion of the chart, indicating that they attract a bigger audience compared to others.  
- This suggests that content type plays an important role in gaining subscribers on YouTube.

- On the other hand, smaller slices represent categories with less overall reach, which may indicate niche audiences or lower popularity.

## Channel Creation Years

- In this section, I analyze the distribution of the years in which YouTube channels were created.

- A box plot is used to understand the spread of channel starting years and to detect any unusual values or outliers.

```python
fig = px.box(df, y='started', title='Years Started')
fig.show()
```

### Insight

- From the box plot, we can observe how YouTube channels are distributed over time.

- Most channels fall within a certain range of starting years, showing periods of higher platform growth.

- The median value represents the central period when most channels in the dataset were created.

- Any points outside the main range can be considered outliers, representing either very early or very recent channels compared to the majority.

## Relationship Between Starting Year and Rank

- In this section, I analyze the relationship between the year a YouTube channel started and its ranking.

- The goal is to see whether older or newer channels tend to have better ranks, and whether category plays a role in this distribution.

```python
fig = px.scatter(df, x='started', y='rank', title='Youtuber per started year and rank', color='category')
fig.show()
```

### Insight

- From the scatter plot, we can observe that there is no perfectly clear linear relationship between starting year and rank.

- Channels from different starting years can appear across various rank positions, meaning that both older and newer channels can achieve high rankings.

- The category coloring also shows that different types of content are distributed across the ranking range, suggesting that category alone does not determine success.

- Overall, ranking seems to depend on multiple factors rather than just the creation year.
